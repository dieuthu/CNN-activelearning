## Show the performance of active learning + cnn
Active learning with most uncertain strategy where the uncertainty is measured with entropy

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('eng-cnn-sep-active-may16.csv')


## Calculate t-test for 
output = df.groupby(['query_strategy', 'number_of_instances']).mean()
x = list(df['number_of_instances'].unique())
lc_mean=list(output.unstack().loc['lc'])
rs_mean = list(output.unstack().loc['rs'])
#lc_std = list(output.unstack().loc['lc'])
#rs_std = list(output.unstack().loc['lc'])

plt.plot(x, rs_mean, c='r')
plt.plot(x, lc_mean, c='b')

# Analyse the variance of the output
output = df.groupby(['query_strategy', 'number_of_instances']).mean()
print (output)
```

## Show the distribution of the output of CNN-SEP
Train CNN-sep with 48 samples, perform prediction on unlabeled data. The output is a distribution of probabilities that a document belongs to 4 level classes. We then perform dimension reduction with TSNE to obtain 2 dimension points for visualization.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('eng-cnn-sep-48trn_unlabeled_tsne_output_may10.csv', names=['a', 'b','c'])
df.replace(1,'r',inplace=True)
df.replace(2,'g',inplace=True)
df.replace(3,'b',inplace=True)
df.replace(4,'y',inplace=True)

df.plot.scatter(x='a',y='b',c=df['c'].values)

```
